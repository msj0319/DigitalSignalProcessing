```python
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
```

```python
wave = thinkdsp.read_wave('piano.wav')

wave.normalize() #범위가 -1에서 1을 초과하지 않도록 웨이브 크기를 조절.
wave.apodize() #웨이브의 시작과 끝을 가늘게하여 재생할 때 클릭하지 않도록 함.
wave.plot() 
thinkplot.config(xlabel='Time (s)')
```

피치가 일정한 0.5초 간격의 세그먼트 선택. 스펙트럼 그리기

```python
start = 2.8
duration = 0.5
segment = wave.segment(start, duration)
segment.plot()
thinkplot.config(xlabel='Time (s)')
segment.make_audio()
```

잘라낸 세그먼트 스펙트럼

```python
spectrum = segment.make_spectrum()
spectrum.plot(high=1000)
thinkplot.config(xlabel='Frequency (Hz)')
```

어떤 스펙트럼도 정형파의 성분으로 나타낼 수 있다

low_pass 스펙트럼을 wave로 변환하여 듣기

```python
spectrum.low_pass(880)
spectrum.plot(high=2000)
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
spectrum.make_wave().make_audio()
```

```python
wave = spectrum.make_wave()
wave.normalize() #범위가 -1에서 1을 초과하지 않도록 웨이브 크기를 조절.
wave.apodize() #웨이브의 시작과 끝을 가늘게하여 재생할 때 클릭하지 않도록 함.
wave.plot() 
thinkplot.config(xlabel='Time (s)')
wave.make_audio()
```
