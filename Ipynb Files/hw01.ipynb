```python
wave = thinkdsp.read_wave('piano.wav')

wave.normalize() # -1 1     .
wave.apodize() #        .
wave.plot() 
thinkplot.config(xlabel='Time (s)')
```

  0.5   .  

```python
start = 2.8
duration = 0.5
segment = wave.segment(start, duration)
segment.plot()
thinkplot.config(xlabel='Time (s)')
segment.make_audio()
```

```python
spectrum = segment.make_spectrum()
spectrum.plot(high=1000)
thinkplot.config(xlabel='Frequency (Hz)')
```

low_pass  wave  

```python
spectrum.low_pass(880)
spectrum.plot(high=2000)
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
spectrum.make_wave().make_audio()
```

```python
wave = spectrum.make_wave()
wave.normalize() # -1 1     .
wave.apodize() #        .
wave.plot() 
thinkplot.config(xlabel='Time (s)')
wave.make_audio()
```

high_pass  wave  

```python
spectrum.high_pass(440)
spectrum.plot(high=2000)
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
wave = spectrum.make_wave()
wave.normalize() # -1 1     .
wave.apodize() #        .
wave.plot() 
thinkplot.config(xlabel='Time (s)')
wave.make_audio()
```

```python
spectrum.band_stop(low_cutoff=440,high_cutoff=880)
spectrum.plot(high=2000)
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
wave = spectrum.make_wave()
wave.normalize() # -1 1     .
wave.apodize() #        .
wave.plot() 
thinkplot.config(xlabel='Time (s)')
wave.make_audio()
```
